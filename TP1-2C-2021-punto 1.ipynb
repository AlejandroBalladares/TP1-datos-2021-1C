```python
df_logs['timestamp'] = pd.to_datetime(df_logs['timestamp'])
df_logs['Dia_de_la_semana'] = df_logs['timestamp'].transform(lambda x: x.day_name())
```

```python
df_respuesta_1 = df_logs.groupby('Dia_de_la_semana')['comment'].agg({'count'})
df_respuesta_1.columns = ['acciones_realizadas']
df_respuesta_1 #acciones realizadas por usuarios según el día de la semana
```
